```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def plot_feature_vectors(df):
    n = 10
    for i in range(n):
        feature_vector = df.iloc[i, :-1].values  # Excluir la columna 'covid'

        # Crear una gr√°fica de barras para visualizar el feature vector
        plt.figure(figsize=(10, 6))
        plt.bar(range(1, len(feature_vector) + 1), feature_vector)
        plt.xlabel('Feature Number')
        plt.ylabel('Feature Value')
        plt.title(f'Feature Vector for Audio {i + 1}')
        plt.show()
```

```python
df = pd.read_csv('audio_features.csv')

plot_feature_vectors(df)
```
